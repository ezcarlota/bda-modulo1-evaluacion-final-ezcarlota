
AGREGAR PRODUCTOS (nombre,precio,cantidad)

```python
inventario = [{"nombre":"camiseta" , "precio": 5 , "cantidad":30}]
```

```python
def agregar_producto(nombre , precio , cantidad): #definicion de la funcion
    
     for producto in inventario:
      
          if producto["nombre"]. lower() == nombre.lower(): # Normalizo texto por ahorrame posibles errores de escritura
             producto["cantidad"] += cantidad # actualizacion de cantidad pongo el + porque si un producto existe quiero agragar cantidad
             print(f"Cantidad de {nombre} actualizada es {producto['cantidad']}") 
             return #salimos de la funcion al encontrarlo 
        
        #si el bucle termina sin encotrar el producto lo agregamos 
     inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
     print("Hemos agregado el producto al inventario")
        
    
           
```

```python
agregar_producto("pantalones",30,45)
agregar_producto("abrigo",60,25)
agregar_producto("camiseta",9.95,100)
```

```python
def eliminar_producto(nombre_a_eliminar):# definir funcion
    for producto in inventario:
      if producto["nombre"].lower() == nombre_a_eliminar.lower():
       inventario.remove(producto) #para eliminar 
       print(f"{producto['nombre']} eliminado con éxito")
       return
    #si no encotramos el articulo en el inventario
    print("Este articulo no se encuentra en inventario")
```

ACTUALIZAR_STOCK(nombre,cantidad)

```python
def actualizar_stock(nombre_producto ,nueva_cantidad,precio=None): #definimos la funcion
   
    for producto in inventario:
        if producto["nombre"].lower() == nombre_producto.lower():
           producto["cantidad"] = nueva_cantidad
           if precio is not None:
              producto["cantidad"] = nueva_cantidad
           print(f"Cantidad: {nombre_producto} es {producto['cantidad']}") #imprimimos el valor que habia en en invetario
           return
    #se añade una nueva cantidad a producto
    precio_nuevo = precio if precio is not None else 0
    inventario.append({"nombre": nombre_producto , "cantidad": nueva_cantidad,"precio": precio_nuevo})
    print(f"Producto'{nombre_producto}' agregado al inventario con cantidad {nueva_cantidad} producto creado por no encontrase en el inventario")
   
        
   

    
    
```

CALCULAR VALOR INVENTARIO


```python
def calcular_valor_inventario(inventario, operacion="multiplicacion"):
    total_acumulado = 0
    
    for producto in inventario:
        # Extraemos los valores del diccionario
        precio = producto.get("precio", 0)
        cantidad = producto.get("cantidad", 0)
        
        if operacion.lower() in ["multiplicacion", "*"]:
            valor_producto = precio * cantidad
            print(f"Producto: {producto['nombre']} | Valor: {valor_producto}")
            total_acumulado += valor_producto
    return total_acumulado #retorna la suma de todos lo productos
```

```python
calcular_valor_inventario(inventario, operacion="multiplicacion")
```

REALIZAR COMPRA


```python
inventario = [{'nombre': 'pantalones', 'precio': 30, 'cantidad': 20},
 {'nombre': 'abrigo', 'precio': 60, 'cantidad': 25},
 {'nombre': 'camiseta', 'cantidad': 130, 'precio': 5.95},
 {'nombre': 'bañador', 'cantidad': 12, 'precio': 35}]

def realizar_compra(inventario):
    carrito = [] 
    total = 0
    
    while True:
        print("---------------------------")
        print("Inventario disponible")
        print("---------------------------")
        for producto in inventario:
            print(f"{producto['nombre']},{producto['precio']}€ , stock:{producto['cantidad']} " )
             
        eleccion = input("¿Que estas buscando?('o salir'):"). lower().strip()
        print(f"Has añadido {eleccion} a tu carrito")
        if eleccion =="salir":
            break

        encontrado = False

        for producto in inventario:
            if producto["nombre"] == eleccion:
                encontrado = True
                if producto["cantidad"] > 0:
                   producto["cantidad"]-= 1
                   carrito.append(producto["nombre"]) 
                   total += producto["precio"]
                   print("Producto añadido al carrito")  
            
                else:
                   print("No hay stock")
        if not encontrado:
            print("Ese producto no existe")
    print("-----------------------")
    print("\nCOMPRA EN EL CARRITO")
    print("-----------------------")
    print("Carrito",carrito)
    print(f"TOTAL A PAGAR : {total:.2f} €")# convertimos los nuemeros en foat con el .2f
    print("Muchas gracias por su compra \n HASTA PRONTO") 
```

```python
realizar_compra(inventario)
```

```python

clientes = [] #utilizo una lista que puede contener varios diccionarios 
def agregar_cliente(nombre,email):
    
    for cliente in clientes:
       if cliente["nombre"].lower() == nombre.lower():
        print(f"Clienta en base de datos: {nombre} | email:{cliente['email']} ")
        return
        
    clientes.append({
     "nombre": nombre,  
      "email": email
      })
    print("Clienta agregada con exito")

```
